## Conversão de tipos

```
def verificacao_dirigir_sem_parametro():
  idade = input('Qual sua idade? ')
  idade = int(idade)
  if idade >= 18:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')
verificacao_dirigir_sem_parametro()
```

```
def verificacao_dirigir(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, tem permissão para dirigir')
  else:
    print(f'{idade} anos de idade, não tem permissão para dirigir')
for idade in idades:
  verificacao_dirigir(idade)
```

```
def verificacao_dirigir(idade):
  for idade in idades:
    if idade >= 18:
      print(f'{idade} anos de idade, tem')
    else:
      print(f'{idade} anos de idade, não')
  
verificacao_dirigir(idade)
```

## Tipo Booleano

```
idade = 15
idade >= 18
```

```
permissoes = []
idades = [10, 14, 50]

def verificacao_dirigir(idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

verificacao_dirigir(idades, permissoes)
```

```
permissoes
```

```
for permissao in permissoes:
  if permissao == True:
    print('Tem')
  else:
    print('Não')
```
